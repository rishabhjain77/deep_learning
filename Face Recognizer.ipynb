```python
import cv2
import numpy as np
```

```python
faceDetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
cam = cv2.VideoCapture(0)
rec=cv2.face.LBPHFaceRecognizer_create()
rec.read("recognizer//trainingData.yml")
id = 0
while(True):
    ret,img = cam.read() #ret - whether cam is working ,if yes return img
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = faceDetect.detectMultiScale(gray,1.3,5)
    for(x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        id,config = rec.predict(gray[y:y+h,x:x+w])
        if id==1:
            name="rishabh"
        elif id==2:
            name = "Berlin"
        org = (x,y+h) #where to display result
        font = cv2.FONT_HERSHEY_COMPLEX #font of text
        fontscale = 1 #size of text
        color = (255,0,0) #color of text
        thickness = 2
        cv2.putText(img,name,org,font,fontscale,color,thickness)
    cv2.imshow("Face",img)
    if(cv2.waitKey(1)==ord('q')):
        break
cam.release()
cv2.destroyAllWindows()
```
